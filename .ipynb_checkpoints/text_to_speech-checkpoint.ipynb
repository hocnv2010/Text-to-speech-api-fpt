## Text to speech use FPT  API
#### Step 1: create file input.txt
#### Step 2: run all lines

```python
#Text to speech use fpt api
import requests

api_key = "7c1151e8d66146cc832ba10584a5333a"
voice = "male"
```

```python
from textwrap import wrap
import time
import wget
file = open("input.txt", "r", encoding="utf-8") 
content = file.read()

wraptexts = wrap(content, 480)

for i in range(len(wraptexts)):
    while True:
        try:
            text = wraptexts[i]
            url = "http://api.openfpt.vn/text2speech/v4?api_key={}&voice={}".format(api_key, voice)
            response = requests.post(url, data=text.encode('utf-8'))
            response = response.json()
            print('\n', response['async'])
            file = response['async']
            print("downloading file {}/{} ".format(i+1, len(wraptexts)), "{}.mp3".format(i))
        except :
            time.sleep(1)
            continue
        break
    while True:
        try:
            wget.download(file, "{}.mp3".format(i))
        except :
            time.sleep(1)
            continue
        break
    
print('\nCOMPLETE')
    
```
